```python
#import necessary libraries
import matplotlib.pyplot as plt 
import numpy as np

#define relevant functions
def f(x):
      return x**2 
def g(x):
    return (x**4)-(2*x**2)
def funny_function(x):
    if x>0:
        return x**x
    elif x==0:
        return 1
    else:
        return abs(x)**abs(x)
def deriv(f, base):
      return ((f(base+10**(-5))-f(base))/(10**(-5)))
def abs_x(x):
    return np.abs(x)



#define difference between very small step to the left and right
def gradient_descent(f, L, x0):
    x_coordinates = []
    y_coordinates = []
    x_coordinates.append(x0) #establish first x coordinate as x0
    y_coordinates.append(f(x0)) #establish first y coordinate as f(x0)
    xi=x0
    i=1
    while abs(deriv(f,xi))>0.01:
        xi=x_coordinates[i-1]-L*deriv(f,x_coordinates[i-1])
        #x1 is x0 + learning_value * f'(x) at this point
        yi=f(xi)
        x_coordinates.append(xi) #add xi to x as the last value
        y_coordinates.append(yi) #add yi to y as the last value
        i=i+1



    plot_range = np.linspace(min(x_coordinates)-0.5,max(x_coordinates)+0.5,1000)
    f_range = [f(i) for i in plot_range]
    plt.plot(plot_range,f_range) #plot the range and function
    plt.plot(x_coordinates,y_coordinates) #plot the x and y coords of Euler's method
    plt.show()
    print(x_coordinates[i-1])
    return round (x_coordinates [-1],3), round (y_coordinates [-1], 3) #returns your
    # last an and yn, #rounded to three decimal placed.

gradient_descent(f,0.9,5)
gradient_descent(g, 0.1, 0.7)
gradient_descent(g, 0.1, -1.7)
gradient_descent(funny_function, 0.1, -1)
gradient_descent(funny_function, 0.1, 1.7)
```
