```python
import sympy as sp

#a
x, y = sp.symbols('x, y')
f = x**2 + sp.sin(y)
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

#b
def gradient_vector(x, y):
    return (df_dx, df_dx)

print(gradient_vector(1, -1))
#gradient for single variable;

#c
h = sp.log(x**2+y**2)
dh_dx = sp.diff(h, x)
dh_dy = sp.diff(h, y)
dh2_dx2 = sp.diff(dh_dx, x)
dh2_dy2 = sp.diff(dh_dy, y)
df2_dyx = sp.diff(dh_dy, x)
df2_dxy = sp.diff(dh_dx, y)
print(dh_dx)
print(dh_dy)
print(dh2_dx2)
print(dh2_dy2)
print(df2_dyx)
print(df2_dxy)

#d
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3

j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
#function of my choice? k

x, y = sp.symbols('x y')
k = sp.exp(-x**2 - y**2)

k_func = sp.lambdify((x, y), k, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = k_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $k(x, y) = exp(x*y)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
