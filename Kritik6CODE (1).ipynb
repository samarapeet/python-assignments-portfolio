```python
#import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
def fx(mu, sig, x):
    return ((2*np.pi*(sig))**-0.5)*np.exp(-((x-mu)**2)/(2*(sig))) # density function as given in the assignment
def integration(fx,a,b):
    n=10000 # number of iterations
    x=np.linspace(a,b,n) # make n values for x, evelnly spaced to integrate
    y_middle = []
    delta_x = (b-a)/n # x intervals
    for i in range(n-1):
        xm = (x[i]+x[i+1])/2 # determines the midpoint 
        ym = fx(171,7.1**2, xm) # st.dev and mean for the average male height, for f(x) of xi
        y_middle.append(ym) # sum x and y values in the x interval
    ms = sum(y_middle)*delta_x #compute the integral
    return ms
    
print("The integral of the normal density is", integration(fx,162,190))
x = np.linspace(-10,10,100)
fx_values = fx(0,5,x)
#print the normal density function
plt.plot(x,fx_values, label = 'normal distribution')
plt.show()
```
